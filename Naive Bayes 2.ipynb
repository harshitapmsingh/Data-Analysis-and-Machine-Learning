```python
import pandas as pd
df=pd.read_csv("/Users/harshit_mac/Desktop/datascience/spam.csv")
df.head(5)
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
df
```

```python
df.drop(['Category'],axis='columns',inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(df.Message,df.spam,test_size=0.2)
```

```python
#now we want to convert these Message from character string to Numerals
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
X_train_count=cv.fit_transform(X_train.values)
X_train_count
```

```python
cv.get_feature_names()
```

```python
X_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
```

```python
email_count=cv.transform(emails)
```

```python
email_count
```

```python
email_count.toarray()
```

```python
model.predict(email_count)
```

```python
X_test_count=cv.transform(X_test)
```

```python
model.score(X_test_count,y_test)
```

```python
#till now for every text we were first transforming it using CountVectorizer and then feeding the countmatrix to the
#model.In several cases we wight have to apply different transformations before feeding our data into model.Hectic task
#This can be solved using a pipeline consisting of the steps the data has to be operated on.

#sklearn has a ''''pipeline'''' feature that helps define a pipeline for the incoming data to do transformations.
```

```python
from sklearn.pipeline import Pipeline
cvmp=Pipeline([
    ('vector',CountVectorizer()),
    ('mnb',MultinomialNB())
])
```

```python
cvmp.fit(X_train,y_train)
```

```python
cvmp.score(X_test,y_test)
```

```python
cvmp.predict(emails)
```

```python
from sklearn.datasets import load_wine
wine=load_wine()
```

```python
df=pd.DataFrame(wine.data,columns=wine.feature_names)
```

```python
df['class']=wine.target
```

```python
df
```

```python
df.columns.isna().any()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(df.drop(['class'],axis='columns'),df['class'],test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
model1=GaussianNB()
model2=MultinomialNB()
```

```python
model1.fit(X_train,y_train)
```

```python
model2.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

```python
model2.score(X_test,y_test)
```
