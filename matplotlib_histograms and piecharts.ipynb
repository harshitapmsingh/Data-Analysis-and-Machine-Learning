```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
bloodsugar_level=[88,85,133,147,123,111,84,65,69,143]
plt.hist(bloodsugar_level)
```

```python
#use bins to set no of intervals
bloodsugar_level=[88,85,133,147,123,111,84,65,69,143]
plt.hist(bloodsugar_level,bins=5)
```

```python
plt.hist(bloodsugar_level,bins=5,rwidth=0.95)
```

```python
bloodsugar_levelmen=[88,85,133,147,123,111,84,65,69,143]
bloodsugar_levelwomen=[84,98,100,122,91,153,126,55,117,101]
plt.hist([bloodsugar_levelmen,bloodsugar_levelwomen],bins=[50,80,100,130,180],rwidth=0.95,color=['green','blue'],label=['men','women'])
plt.legend()
```

```python
bloodsugar_levelmen=[88,85,133,147,123,111,84,65,69,143]
bloodsugar_levelwomen=[84,98,100,122,91,153,126,55,117,101]
plt.hist([bloodsugar_levelmen,bloodsugar_levelwomen],bins=[50,80,100,130,180],rwidth=0.95,color=['green','blue'],label=['men','women'],orientation='horizontal')
plt.legend()
```

```python
exp_val=[120,80,30,100]               #here list is used but we can also use pandas dataframes,numpy arrays too
exp_name=['cig','food','shake','misc']       #pie values start at right horizontal level and go anti-clockwise
plt.pie(exp_val,labels=exp_name)
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,radius=1.5,startangle=45)
plt.show()
```

```python
#use autopct for percentages of values shown also specify the places of decimal points you want
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%')
plt.show()
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0,0,0])
plt.show()
```

```python
#use explode to highlight a indexed pie
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0.2,0,0])
plt.show()
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0.2,0,0.2])
plt.show()
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0,0,0])
plt.savefig('first.png')
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0,0,0])
plt.savefig('first.png',bbox_inches="tight")
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0,0,0])
plt.savefig('first.png',bbox_inches="tight",pad_inches=1.5)          
```

```python
exp_val=[120,80,30,100]
exp_name=['cig','food','shake','misc']
plt.pie(exp_val,labels=exp_name,autopct='%0.3f%%',explode=[0,0,0,0])
plt.savefig('first.png',transparent=True)          #setting background as transparent so that we can use it on some websites,etc.
```
