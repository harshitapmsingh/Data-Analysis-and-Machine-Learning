```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
digits.images[0]
```

```python
plt.gray()
for i in range(10):
 plt.matshow(digits.images[i])
```

```python
digits.target[0:7]
```

```python
#handwritten data (like in the images above) will be our input and target are the values that will be predicted
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(x_train)  
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
#the model works well for most cases but for cases where it fails and to get an idea of your model's accuracy.
# use CONFUSION MATRIX

#A confusion matrix is a table that is often used to describe the performance of a classification model 
#(or "classifier") on a set of test data for which the true values are known.
#for more visit: https://towardsdatascience.com/understanding-confusion-matrix-a9ad42dcfd62
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
