```python
import pandas as pd
r_df1=pd.DataFrame(
{
    'city':['mumbai','new york','paris','orlando'],
    'temperature':[32,50,42,51],
    
})
r_df2=pd.DataFrame(
{
    'city':['new york','mumbai','seoul'],
    'windspeed':[16,11,13],
})
```

```python
# using merge tracks the index of values in specified column automatically
#    by default the operation is the intersection or inner join of the two sets
dfg=pd.merge(r_df1,r_df2,on='city')
dfg
```

```python
#using 'how' specifier we can produce a dataframe containing outer join
dfh=pd.merge(r_df1,r_df2,on='city',how='outer')
dfh
```

```python
dfh1=pd.merge(r_df1,r_df2,on='city',how='left')                     #left join
dfh1
```

```python
dfh2=pd.merge(r_df1,r_df2,on='city',how='right')                    #right join
dfh2
```

```python
#indicator showing which dataset the data has come from
dfh=pd.merge(r_df1,r_df2,on='city',how='outer',indicator=True)  
dfh
```

```python
rdf1=pd.DataFrame(
{
    'city':['mumbai','new york','paris','orlando'],
    'temperature':[32,50,42,51],    
})
rdf2=pd.DataFrame(
{
    'city':['new york','mumbai','orlando','paris'],
    'temperature':[48,39,45,39],    
})
```

```python
#for the same city name different temp values of two datasets are recorded in different columns
dfr=pd.merge(rdf1,rdf2,on='city',how='outer')
dfr
```

```python
dfr=pd.merge(rdf1,rdf2,on='city',how='outer',suffixes=('_first','_second'))
dfr
```
