```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[22,56,45,33,87,21,65,76,75]
```

```python
plt.plot(x,y)
```

```python
#matplotlib api plot for extra specifications
plt.plot(x,y,color='magenta',linewidth=5,linestyle=':')
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(x,y)
```

```python
#using format strings(color,marker,line)like here bo--
plt.plot(x,y,'bo--')
```

```python
#see plot function documentation
plt.plot(x,y,'y|:')
```

```python
plt.plot(x,y,'bo')
```

```python
#order doesn't matter
plt.plot(x,y,'ob')
```

```python
#you can use your own color in hexadecimal notation
plt.plot(x,y,color='#004810',linestyle='--',marker='o',markersize=20)
```

```python
#alpha range[0,1] to control transparency
plt.plot(x,y,'bo--',alpha=1)
```

```python
plt.plot(x,y,'bo--',alpha=0.2)
```
