```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company=['google','facebook','amazon','microsoft']
revenue=[11,8,10,7]
profit=[4,3,4,2]
plt.bar(company,revenue)
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
#.  xticks(),yticks() 
#.  matplotlib.pyplot.xticks(ticks=None, labels=None, **kwargs) 
#.  Get or set the current tick locations and labels of the axis. 
#Calling this function with no arguments (e.g. xticks()) is the pyplot equivalent of calling get_xticks and 
#get_xticklabels on the current axes. Calling this function with arguments is the pyplot equivalent of calling 
#set_xticks and set_xticklabels on the current axes.
```

```python
plt.xticks(xpos,company)                    
plt.bar(xpos,revenue)
```

```python
plt.ylabel('revenue in usd billions')
plt.title('company revenues in 2019')
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
#such bar chart was possible without xpos too
plt.ylabel('revenue in usd billions')
plt.title('company revenues in 2019')
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
plt.bar(xpos,profit)
```

```python
#this kind of chart not possible without xpos
plt.ylabel('revenue in usd billions')
plt.title('company revenues in 2019')
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue)
plt.bar(xpos+0.2,profit)
```

```python
#role of xpos in arraanging different types of values for single items
plt.ylabel('usd billions')
plt.title('company revenues in 2019')
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='profits')
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.barh(xpos,revenue)
plt.barh(xpos,profit)
```
