```python
import pandas as pd
df=pd.read_csv("/Users/harshit_mac/Desktop/datascience/salaries.csv")
df
```

```python
inputs=df.drop('salary_more_then_100k',axis='columns')
target=df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
inputs['company_d']=le_company.fit_transform(inputs['company'])
inputs['job_d']=le_job.fit_transform(inputs['job'])
inputs['degree_d']=le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs=inputs.drop(['company','job','degree'],axis='columns')
```

```python
inputs
```

```python
from sklearn.model_selection import train_test_split
inputs_train,inputs_test,target_train,target_test=train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_train,target_train)
```

```python
model.score(inputs_train,target_train)
```

```python
model.predict([[1,1,1]])
```
